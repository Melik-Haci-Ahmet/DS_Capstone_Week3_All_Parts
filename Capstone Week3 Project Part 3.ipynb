## Entering the url in cell below:

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
data=requests.get(url).text
```

```python
html_data_parsed = BeautifulSoup(data,"html5lib")
```

```python
for row in html_data_parsed.find("tbody").find_all("tr"):
    col = row.find_all("td")
#print(html_data_parsed.prettify())
```

```python
TwoTables=html_data_parsed.find_all("table")
#print(TwoTables[0]) #first column is the table that we are looking for.
```

Below we fill our dataframe table with a for loop.

```python
info_table = pd.DataFrame(columns=["Postal Code", "Borough", "Neighbourhood"])
for row in TwoTables[0].find("tbody").find_all("tr"):
    col = row.find_all("td")
    if len(col)>0 : #Because we observed that first index  of col is null.
        p_code = col[0].text
        boro = col[1].text
        neigh = col[2].text
    else: continue
    #print(p_code, boro, neigh)
    info_table = info_table.append({"Postal Code":p_code, "Borough":boro, "Neighbourhood":neigh}, ignore_index=True)
info_table['Postal Code']=info_table['Postal Code'].replace({'\\n':''},regex=True) 
info_table['Borough']=info_table['Borough'].replace({'\\n':''},regex=True)
info_table['Neighbourhood']=info_table['Neighbourhood'].replace({'\\n':''},regex=True)
info_table        
```

Now, we need to drop rows with 'Not assigned' value. 
Doing it for the Borough column also takes care of the Neighbourhood columns as well. 
And also we need to reset the index of the dataframe.

```python
info_table['Borough'].replace('Not assigned', np.nan, inplace=True)
info_table.dropna(subset=['Borough'],inplace=True)
info_table=info_table.reset_index(drop=True)
info_table.tail()
```

First, Let's discover how many boroughs have 'Toronto' as follows.

```python
i=0
for word in info_table['Borough']:
    if 'Toronto' in word.split():
        #print(word) 
        i=i+1
print('There are',i, 'out of 103 Boroughs that contain word Toronto.')
```

New dataframe is created and filled with the respective boroughs as below:

```python
tor_df = pd.DataFrame(columns=["Postal Code", "Borough", "Neighbourhood","Latitude","Longitude"])
for i in info_table['Borough'].index:
    if 'Toronto' in info_table['Borough'][i].split():
        tor_df = tor_df.append({"Postal Code":info_table['Postal Code'][i], 
                                "Borough":info_table['Borough'][i],
                                "Neighbourhood":info_table['Neighbourhood'][i],
                                "Latitude":info_table['Latitute'][i],
                                "Longitude":info_table['Longitude'][i]}, ignore_index=True)
tor_df
```

As We have the dataframe of neighborhoods and boroughs ready, we can plot them via Folium.

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium
print('Necessary library for mapping is installed')
```

Now we can create below the neighbourhoods map of Toronto just as we did in the NYC notebook.

```python
map_toronto = folium.Map(location=[tor_df['Latitude'][0], tor_df['Longitude'][0]], zoom_start=12)

for lat, lng, borough, neighborhood in zip(tor_df['Latitude'], tor_df['Longitude'], tor_df['Borough'], tor_df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
map_toronto
```
